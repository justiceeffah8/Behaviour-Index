```python
#Enter the Fear and Greed Index
import math  # Import the math module to use the log function

print('Enter the Fear and Greed indexs' )
print('=================================')
x1 = float(input('Enter the value of American stock market Fear and Greed index ') )
x2 = float(input('Enter the value of Bitcoin Fear and Greed index ' ) )
print('==================================')


#Compute the Behaviour Index
BI= (math.log(x1)+math.log(x2))/(8)

print('The Behaviour Index is %0.2f' % BI)
```
